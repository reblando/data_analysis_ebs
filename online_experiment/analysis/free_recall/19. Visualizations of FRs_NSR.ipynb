## 2. Function for normalizing vectors

```python
def norm_300(matrix):
    n, m = matrix.shape
    for i in range(0,n):
        sum_ = 0
        for j in range(0, 300):
            sum_ = matrix[i][j] * matrix[i][j]
        k = math.sqrt(sum_)
        for j in range(0,300):
            matrix[i][j] = matrix[i][j]/k
    return matrix
    
```
