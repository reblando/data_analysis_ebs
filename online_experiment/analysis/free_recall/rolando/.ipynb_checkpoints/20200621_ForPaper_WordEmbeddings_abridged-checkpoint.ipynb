useful functions and variables

```python
###
###  VARIABLES
###


subjects = ['AB',  'AC', 'AF', 'BD', 'BO', 'CM', 'EC', 'GN', 'JA',  #Note CM2 removed
            'JA2', 'JF', 'JH', 'JL', 'JP', 'KB', 'LZ',  'MA', 'MD', 'MG',
            'MM',  'MS', 'QH', 'RH', 'RW', 'ST', 'TB',  'TG', 'TK', 'TW', 'VR']

stories = ['Brazil',  'Derek', 'MrBean',   'PulpFiction',
           'BigBang', 'Santa', 'Shame',    'Vinny',
           'DueDate', 'GLC',   'KandD',    'Nonstop',
           'Friends', 'HIMYM', 'Seinfeld', 'UpInTheAir']

nPerm = 1000
nStories = 16 
modality = 'VVVVAAAAVVVVAAAA'
schema_modality = 'AAAABBBBCCCCDDDD'
sc_order = 'RRRRRRRRAAAAAAAA' # here its 16 stories askjdf 18 stories includes airport and restuarnt schema

###
### MASK
###


MASK = np.zeros((nStories,nStories)) #single story schema score mask

for st in range(nStories):
    curr_schema = sc_order[st]
    
    train_ind = [sc==curr_schema for sc in sc_order]*(np.arange(nStories)!=st) 
    test_ind = np.invert(train_ind)*(np.arange(nStories)!=st) 
    
    MASK[st,st] = 1 ## diagonal squares
    MASK[st,train_ind] = 2 ## same schema (top left & bottom right quadrants)
    MASK[st,test_ind] = 3 ## diff schema|offdiag (bottom left & top right qudrant)
    
    
###
### STORY / SCHEMA SCORING 
###

def GetStorySchemaEffect(smat,nPerm=nPerm,MASK=MASK,modality=modality,schema_modality=schema_modality,seed=None):
    '''
        -similarity matrix story and schema extractor function.
        -Calculate Story and Schema Effect from similarity matrix.
        -smat.shape should be (nStories, nStories, nSubj)
    '''
    
    ## 
    assert MASK.shape[0] == smat.shape[0] == len(modality) == len(schema_modality)
    
    nStories = smat.shape[0]
    nSubj = smat.shape[2]

    story_effect = np.zeros((nSubj,nStories,nPerm+1))
    schema_effect = np.zeros((nSubj,nStories,nPerm+1))
    
    # set a seed for reproducibility if provided in arguments
    np.random.seed(seed) if seed != None else None
    
    for s in (range(nSubj)):

        ## prepare 16 story x 16 story similiarity matrices for permutation
        smat_perm_story = smat[:,:,s] 
        smat_perm_schema = smat[:,:,s]

        for p in range(nPerm+1):
            for st in range(nStories):

                subset_mask = [modality[st]==mod for mod in modality]  # index into subsets of the 16 stories

                ### 
                ### STORY EFFECT 
                ###
                diag_square = smat_perm_story[st,st] # story in question
                same_schema_ind = np.logical_and(MASK[st,:]==2,subset_mask)
                avg_same_schema = (np.nanmean(smat_perm_story[st,same_schema_ind]) + 
                                  np.nanmean(smat_perm_story[same_schema_ind,st])) * (1/2)
                
                # get similarity measure for story effect
                story_effect[s,st,p] = diag_square - avg_same_schema

                #visualize
    #             x = np.zeros((16,16))
    #             x[st,same_schema_ind] = 1
    #             x[same_schema_ind,st] = 2
    #             plt.imshow(x);plt.show()

                ### 
                ### SCHEMA EFFECT 
                ###    

                ### SCORING WITH MODALITY SENSITIVTY FIX (feb 2020)
                same_schema_ind = np.logical_and(MASK[st,:]==2, subset_mask)
                diff_schema_ind = np.logical_and(MASK[st,:]==3, subset_mask)

                same_schema = (np.nanmean(smat_perm_schema[st,same_schema_ind]) + 
                               np.nanmean(smat_perm_schema[same_schema_ind,st])) * (1/2)

                diff_schema = (np.nanmean(smat_perm_schema[st,diff_schema_ind]) +
                               np.nanmean(smat_perm_schema[diff_schema_ind,st])) * (1/2)   
                
                # get similarity measure for schema effect
                schema_effect[s,st,p] = same_schema - diff_schema

            nextperm = perm_groups(schema_modality) ## 20200215 onwards, to account for the modality sensitivty
            smat_perm_story = smat[nextperm, :, s].copy()


            nextperm = perm_groups(modality)
            smat_perm_schema = smat[np.ix_(nextperm,nextperm)].copy()
            
    return story_effect,schema_effect

###
###  ADDITIONAL FXNS
###
def charmatch(s, c):
    return np.array([c == x for x in s])

def perm_groups(labels):
    groups = ''.join(set(labels))
    perm = np.zeros(len(labels), dtype=np.int_)
    for g in groups:
        group_inds = np.where(charmatch(labels, g))[0]
        perm[group_inds] = np.random.permutation(group_inds)

    return perm

def saveObject(strName, variable):
    with open(strName, 'wb') as handle:
        pickle.dump(variable, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
def openObject(strName):
    with open(strName, 'rb') as handle:
        objectOut = pickle.load(handle)
    return objectOut
```

word embedding preprocessing (a lot of character encoding garbage in transcripts)

```python
wv_dim = 300

# FastText preprocessing, based on bittlingmayer/ft_wiki_preproc.py
# Remove special characters, put spaces between all tokens
SUB = ["s/’/'/g", "s/′/'/g", "s/''/ /g", "s/'/ ' /g", 's/“/"/g', 's/”/"/g', 's/"/ /g', "s/\\./ \\. /g", "s/<br \\/>/ /g", "s/, / , /g", "s/(/ ( /g", "s/)/ ) /g", "s/\\!/ \\! /g", "s/\\?/ \\? /g", "s/\\;/ /g", "s/\\:/ /g", "s/-/ - /g", "s/=/ /g", "s/=/ /g", "s/*/ /g", "s/|/ /g", "s/«/ /g", 
       "s/…/ /g", "s/‘/ /g", "s/í/ /g", "s/ñ/ /g", "s/\x84/ /g", "s/î/ /g", "s/ó/ /g", "s/\x83/ /g", "s/ï/ /g", "s/õ/ /g",
       "s/ò/ /g", "s/,/ /g", "s/ô/ /g", "s/\x92/ /g", "s/é/ /g", "s/\x8e/ /g", "s/â\x80¦/ /g", "s/\x91/ /g", "s/\x93/ /g",
       "s/\x94/ /g", "s/ã®/ /g", "s/ã¨/ /g", "s/ã©/ /g",
       "s/\â\x80\x99/ /g", "s/â\x80\x9c/ /g", "s/â\x80\x9d/ /g", "s/â\x80\x99/ /g", "s/â\x80\x9c/ /g", "s/â\x80\x98/ /g",
       "s/â/ /g"]

def __normalize_text(s):
    for sg in SUB:
        rep = sg.replace('\\','').split('/')
        s = s.replace(rep[1], rep[2])
    s = s.replace('/',' ')
    return s

def __spaces(s):
    return ' '.join(s.split())

def __digits(s):
    return ''.join(filter(lambda c: not c.isdigit(), s))

# def preproc(s):
#     return __punctuation(__spaces(__digits(__normalize_text(s.lower()))))

def preproc(s):
    return (__spaces(__digits(__normalize_text(s.lower()))))

def __punctuation(s):
    return s.translate(str.maketrans('', '', string.punctuation))

def word2vecSent(sentence, model = 'fasttext'):
    wv_dim = 300 #for glove and fasttext
    
    if model == 'glove':
        wvmodel = glove_model
    elif model == 'fasttext':
        wvmodel = wv_model
        
    words = preproc(sentence).split(' ')
    wv = np.zeros((len(words), wv_dim))
    for i in range(len(words)):
        if words[i] in wvmodel.vocab:
            wv[i,:] = wvmodel.word_vec(words[i])
#             wv[i,:] = get_word_emb([words[i]])
    
    return wv

# def word2vecSent(sentence):
#     ## for bert
#     wv_dim = 1024
#     wv = np.zeros((1, wv_dim))
#     bert_sent = bert_embedding([sentence])
#     wv[0,:] = np.array([bert_sent[0][1][i] for i in range(len(bert_sent[0][1]))]).mean(0)
# #     wv[0,:] = get_word_emb([sentence])
#     return wv
```

make word embeddings from subject recalls

```python
wv_dim = 300 #1024
nSubj = 30
subj_recall = np.zeros((nSubj,nStories,wv_dim))
for s, subj in enumerate((subjects)):
    
    print(s,end=',')
    
    for st, story in enumerate(stories):
        
        sentences = T[subj]['Stories'][story]['Sentence']
        nSent = len(sentences) 
        story_sentences = np.zeros((nSent, wv_dim)) #(number of sent, 300)
        
        for sent_i,sent in enumerate(sentences):
#             print(sent)
            ### create SENTENCE vector from averaging the words in a sentence
            story_sentences[sent_i,:] = word2vecSent(sent).mean(0)#/np.linalg.norm(word2vecSent(sent))
                    
        ### average the sentences to get an overall STORY vector for each subject
        subj_recall[s,st,:] = story_sentences.mean(0)
           
print(subj_recall.shape)
```

explore clustering in embedding space 

```python
## average across subjects to get 16 different story vectors
stories_mean = subj_recall.mean(0)

## PCA for first 2 PCs
pca = PCA(n_components=2)
pca.fit(stories_mean)
pc12 = pca.transform(stories_mean)

## plot
plt.figure(figsize=(10,5))

y = ['r' if sc == 'R' else 'b' for sc in sc_order] ## for coloring points by schema

plt.scatter(pc12[:,0], pc12[:,1],c=y)

for i in range(nStories):
    plt.text(pc12[i, 0], pc12[i, 1], s=stories[i] )
    
plt.xlabel("PC 1", fontsize=15); plt.ylabel("PC 2",fontsize=15);


```

prepare similarity matrix and run story and schema scores in embedding space

```python
## center-mean embeddings
centered_recall = subj_recall - subj_recall.mean(0).mean(0)

## prepare similarity matrix via leave-one-subject-out correlations
smat=np.zeros((nStories,nStories,nSubj))
for loo_s in range(nSubj):
    
    loo_vec = centered_recall[loo_s,:,:]
    group_vec = centered_recall[np.arange(nSubj)!=loo_s,:,:].mean(0) #average out subj dimension
    
    smat[:,:,loo_s] = np.corrcoef(loo_vec,group_vec)[:nStories,nStories:]
    
plt.imshow(smat.mean(2));plt.colorbar();
plt.xlabel('subj',fontsize=17);plt.ylabel('group',fontsize=17);
```

run story and schema score measure

```python
### RUN STORY / SCHEMA SCORE
story_score,schema_score = GetStorySchemaEffect(smat[:,:],nPerm=100)
```

### current analysis

collect subject schema descriptions

```python

### COLLECT SUBJECT SENTENCES describing a SCHEMA and then convert them to word embeddings

subj_schema = np.zeros((nSubj,2, wv_dim))

## airport is index 0

for s, subj in enumerate((subjects)):

    sentences = T[subj]['Schemas']['Sentence']

    sch, sch_count = np.unique(T[subj]['Schemas']['Schema'],return_counts=True) #collect number of sentences used for describing each schema
    air_idx = np.array(T[subj]['Schemas']['Schema'])=='Going to an airport' #consistent schema sentence matching per subject
    
    airport_sentences = np.zeros((sch_count[1], wv_dim)) 
    restaurant_sentences = np.zeros((sch_count[0], wv_dim))
    
    ### CONVERT each sentence into an embedding 
    
    for i,idx in enumerate([air_idx,np.invert(air_idx)]):
        
        for sent_i, sent in enumerate(np.array(sentences)[idx]):
            if i==0:
                aitrport_sentences[sent_i,:] = word2vecSent(sent).mean(0)#/np.linalg.norm(word2vecSent(sent))
            else:
                restaurant_sentences[sent_i,:] = word2vecSent(sent).mean(0)#/np.linalg.norm(word2vecSent(sent))
                
    ### average the sentences to get an overall schema vector for each subject
    subj_schema[s,0,:] = airport_sentences[:,:].mean(0)
    subj_schema[s,1,:] = restaurant_sentences[:,:].mean(0)   
        
```

subject story recall distance from subj schema descriptions ***

```python
from scipy.spatial import distance

centered_recall = subj_recall.copy() - subj_recall.mean(0).mean(0) # CENTER people's story recalls
mod_recall = centered_recall#[:,np.arange(16)!=6]

subj_schema_mod = subj_schema.copy() - subj_schema.mean(0).mean(0) # CENTER people's schema description recalls

subj_dist = np.zeros((30,16))

## leave-one-out correlations and contrasts
for loo_s in range(nSubj):
    
    ##extract subj and N-1 group SCHEMA embeddings
    loo_schema_vec = subj_schema_mod[loo_s]
    group_schema_vec = subj_schema_mod[np.arange(nSubj)!=loo_s].mean(0)
    
    ##extract subj and N-1 group STORY embeddings
    loo_story_vec = mod_recall[loo_s]
    group_story_vec = mod_recall[np.arange(nSubj)!=loo_s].mean(0) #average out subj dimension

    for st, story in enumerate(stories):
        x_air = group_schema_vec[0,:] 
        x_rest = group_schema_vec[1,:]
        y = loo_story_vec[st,:]
        
        #how correlated are the story embeddings to the schema embeddings?
        r_air = pearsonr(x_air,y)[0] 
        r_rest = pearsonr(x_rest,y)[0]
        
        ### contrast
        proportion = r_rest - r_air if sc_order[st] == 'R' else r_air-r_rest
        
        ### save
        subj_dist[loo_s,st] = proportion

    
```

how correlated are hand-labeled detail scores vs automated schema scores?

```python
rubric_scores = openObject("normalizedRubricScores_byStory.pkl")

## story
x,y = subj_dist.ravel(),rubric_scores.ravel()
corr = pearsonr(x,y)[0]

plt.plot([0,.5],[0,.5], lw=3,c='k',alpha=0.4,label = 'y=x')
plt.scatter(x,y,alpha=0.7)
plt.xlabel('schema similarity',fontsize=17);plt.ylabel('rubric',fontsize=17);
plt.title('corr: {:.3f}'.format(corr),fontsize=17);
plt.xlim((-.4,1.1))
plt.ylim((-.4,1.1));
plt.legend();

```
