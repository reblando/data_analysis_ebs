```python
# Concatenating wvs for plotting below
wv_dim = 300
wvs_matrix = np.zeros((8, wv_dim))
order_keys = np.zeros((8,1))
count = 0
for key in templates:
    wvs_matrix[count, :] = templates[key]
    order_keys[count] = key
    count += 1
```

### 2. Concatenate all word vectors in story templates

```python
concat_stories = {}

first = 0

keys = [11, 12, 13, 14, 21, 22, 23, 24, 31, 32, 33, 34, 41, 42, 43, 44]
ranges = [8]



for key in keys:
    for i in range(0, 3):
        if first == 0:
            this_matrix = stories[key][i]
            first = 1
        else:
            this_matrix = np.concatenate((this_matrix, stories[key][i]), axis =0)
    ranges.append((ranges[-1] + sums[key][0]))
```

### 3. Concatenate word vectors of templates and stories

```python
all_wv = np.concatenate((wvs_matrix, this_matrix), axis = 0)
```

## D. Avg participants within each story and plot

```python


first = 0


for key in keys:
    for i in range(0, 3):
        if i == 0:
            this_matrix = stories[key][i]
        else:
            this_matrix = np.concatenate((this_matrix, stories[key][i]), axis =0)
    avg_vec = np.mean(this_matrix, axis = 0)
    avg_vec = avg_vec.reshape(-1, 1)
    avg_vec = avg_vec.T
    if first == 0:
        avg_matrix = avg_vec
        first = 1
    else:
        avg_matrix = np.concatenate((avg_matrix, avg_vec), axis = 0)
```

```python
avg_all_wv = np.concatenate((wvs_matrix, avg_matrix), axis = 0)
```
