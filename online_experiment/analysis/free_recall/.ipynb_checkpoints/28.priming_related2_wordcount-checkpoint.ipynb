## B. wordCount function

```python
# FastText preprocessing, based on bittlingmayer/ft_wiki_preproc.py
# Remove special characters, put spaces between all tokens
SUB = ["s/’/'/g", "s/′/'/g", "s/''/ /g", "s/'/ ' /g", 's/“/"/g', 's/”/"/g', 's/"/ /g', "s/\\./ \\. /g", "s/<br \\/>/ /g", "s/, / , /g", "s/(/ ( /g", "s/)/ ) /g", "s/\\!/ \\! /g", "s/\\?/ \\? /g", "s/\\;/ /g", "s/\\:/ /g", "s/-/ - /g", "s/=/ /g", "s/=/ /g", "s/*/ /g", "s/|/ /g", "s/«/ /g", 
       "s/…/ /g", "s/‘/ /g", "s/í/ /g", "s/ñ/ /g", "s/\x84/ /g", "s/î/ /g", "s/ó/ /g", "s/\x83/ /g", "s/ï/ /g", "s/õ/ /g",
       "s/ò/ /g", "s/,/ /g", "s/ô/ /g", "s/\x92/ /g", "s/é/ /g", "s/\x8e/ /g", "s/â\x80¦/ /g", "s/\x91/ /g", "s/\x93/ /g",
       "s/\x94/ /g", "s/ã®/ /g", "s/ã¨/ /g", "s/ã©/ /g",
       "s/\â\x80\x99/ /g", "s/â\x80\x9c/ /g", "s/â\x80\x9d/ /g", "s/â\x80\x99/ /g", "s/â\x80\x9c/ /g", "s/â\x80\x98/ /g",
       "s/â/ /g"]

def __normalize_text(s):
    for sg in SUB:
        rep = sg.replace('\\','').split('/')
        s = s.replace(rep[1], rep[2])
    s = s.replace('/',' ')
    return s

def __spaces(s):
    return ' '.join(s.split())

def __digits(s):
    return ''.join(filter(lambda c: not c.isdigit(), s))

# def preproc(s):
#     return __punctuation(__spaces(__digits(__normalize_text(s.lower()))))

def preproc(s):
    return (__spaces(__digits(__normalize_text(s.lower()))))

def __punctuation(s):
    return s.translate(str.maketrans('', '', string.punctuation))


def wordCount(sentence, model = 'fasttext'):        
    words = preproc(sentence).split(' ')
    
    return len(words)
```

# I. Correlate all participants' recalls with each other
## A. concatenate all participants

```python
story_order = [11,12,13,14,21,22,23,24,31,32,33,34,41,42,43,44]

all_wvs = np.zeros((0,300))
all_ids = np.zeros((0,2))
# count how many participants are in each category for labeling
new_story = [0]
new_priming = [0]

for key in story_order:
    story_count = 0
    for i in range(0, 3):
        priming_count = 0
        n = recalls[key][i].shape[0]
        for j in range(0, n):
            #increase counts
            story_count +=1
            priming_count +=1
            all_wvs = np.vstack((all_wvs, recalls[key][i][j]))
            all_ids = np.vstack((all_ids, ids[key][i][j]))
        # Appending new coordinate to new_priming
        this_one = new_priming[-1] + priming_count
        new_priming.append(this_one)
     # Appending new coordinate to new_story
    this_one = new_story[-1] + story_count
    new_story.append(this_one)
        
```

## B. Correlate all frs

```python
n = all_wvs.shape[0]

fr_corr = np.zeros((n,n))

count = 0

for i in range(0, n):
    for j in range(0,n):
        count += 1
        this_coef = np.corrcoef(all_wvs[i,:], all_wvs[j,:])
        fr_corr[i,j] = this_coef[0,1]
        
```

# III. How to Quantify the "Dark Bands" in each story'
## A. Iterate through all participants, sum their correlations, subtract 1, and avg

```python
correl_vals = []

for i in range (0, fr_corr.shape[0]):
    correl_vals.append((np.sum(fr_corr[i, :]) - 1)/315)
```

## B. Plot histogram

```python
plt.hist(correl_vals, bins = 40)
plt.show()
```

## C. Get the bottom 5% of recalls and print them out

```python
bottom = int(np.ceil(315 * .05))

# converting correl_vals to arry
cv = np.array(correl_vals)

#argsort cv and get bottom numbers
b_ind = np.argsort(cv)[0:bottom]

# index all_ids and use stored key as a key for data_dict and print out the recall
for i in range(0, bottom):
    this_key = all_ids[b_ind[i]][0]
    print('\n\n', data_dict[this_key]['recall'][0])
```

## D. Get the top 5% of the recalls and print them out

```python
top = int(np.ceil(315 * .05))

#argsort cv and get bottom numbers
b_ind = np.argsort(cv)[-top:]

# index all_ids and use stored key as a key for data_dict and print out the recall
for i in range(0, bottom):
    this_key = all_ids[b_ind[i]][0]
    print('\n\n', data_dict[this_key]['recall'][0])
```

# IV. Correlating length of recall with avg correlation for each participant
## A. Make an array of the recall lengths organized in the same way as the avg correlation vector

```python
wc = np.zeros((0,1))

for story in story_order:
    for i in range(0,3):
        n_partc = mrecalls[story][i].shape[0]
        for j in range(0, n_partc):
            this_text = mrecalls[story][i][j][0]
            wc = np.vstack((wc, wordCount(this_text)))
```

## B. Plotting

```python
plt.scatter(wc, correl_vals)
plt.xlabel('Number of Words in Recall')
plt.ylabel('Avg Correlation')
plt.title('Correlation as a function of # of Words in a Recall')
```
