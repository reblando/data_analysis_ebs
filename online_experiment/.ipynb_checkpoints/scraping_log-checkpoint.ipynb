```python
import re
import codecs
import glob
import csv
import pandas as pd
```

```python
files = glob.glob("*.log")  # the folder with the logs (relative to this file)
# we'll use a regex for this even though they're ugly and painful:
scanQuote = re.compile(r"'(.*?)'")  # finds any number of words within a quote
```

```python
count_distractor = 1
for filename in files:
    # open log file
    thisFile = codecs.open( filename, 'r', 'utf-8')
    #open a data file to send csv data
    outFilename = "first/" + filename[:-4]+".csv"
    outFile = codecs.open(outFilename, 'w', 'utf-8')
    outFile.write("time, keypress\n")
    #go through each line in the file
    for line in thisFile.readlines():
        if "DATA" in line:  # new trial reset all entries
            first, keypress = line.split("Keydown:")
            time = first.split("\t")[0]
            outFile.write("{},{}".format(time, keypress))
        if "text_34: autoDraw = false" in line:
            time = "here"
            keypress = 33
            outFile.write("{},{}\n".format(time, keypress))
        #recording the distractor words    
        if "text_66: text =" in line:
            time = 'distractor_text_' + str(count_distractor)
            b, keypress = line.split("text = ")
            outFile.write("{},{}".format(time, keypress))
            count_distractor += 1
        #recording the short answer questions
        if "instr: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q1'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_2: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q2'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_3: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q3'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_4: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q4'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_5: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q5'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_6: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q6'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_7: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q7'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        if "instr_8: text =" in line:
            b, keypress = line.split("text = ")
            time = 'Q8'
            keypress = keypress.strip()
            if keypress == "\"default text\"":
                pass
            else:
                outFile.write("{},{}\n".format(time, keypress))
        #short answers
        text_num = ["43", "44", "45", "46", "47", "48", "51", "53"]
        if "text_43: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A1'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_44: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A2'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_45: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A3'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_46: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A4'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_47: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A5'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_48: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A6'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_51: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A7'
            outFile.write("{},{}\n".format(time, keypress))
        if "text_53: text = " in line:
            b, keypress = line.split("text = ")
            time = 'A8'
            outFile.write("{},{}\n".format(time, keypress))
        
    outFile.close()
```

```python
#Reading in the First CSV in folder "First"
for file in files:
    #reading in csv made above
    data = pd.read_csv("first/" + filename[:-4]+ ".csv", error_bad_lines=False, sep=r'\s*,\s*') 
    #dropping Nan values
    data = data.dropna()
    #finding the index of '33' values
    index33 = data.index[data['keypress'] == '33'].tolist()
    #subtracting 1 from each value in the index to find the keypress
    keypresses = [x - 1 for x in index33]
    #changing pd Series into a list
    key_vals = data['keypress'][keypresses].tolist()
    #forming the dataframe that will be exported
    output_pd = pd.DataFrame(key_vals, columns = ['keypresses'])
    #iterating through the 8 short answer questions and answers
    for i in range(1, 9):
        print(data.loc[data['time'] == ('Q' + str(i)), 'keypress'].iloc[0])
        
    
    print(output_pd)

      
    
```
