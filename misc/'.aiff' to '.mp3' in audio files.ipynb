```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.plotly as py
import plotly
import os
import glob
import math
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
import xlsxwriter

#laptop
directory = '/Users/alexreblando/Documents/GitHub/ebs/behavioral_August2019/psychopy experiment/audio_excel_sheets'

filenames = glob.glob(directory+ '/*.xlsx')

dfs = []


for filename in filenames:
    dfs.append(pd.read_excel(filename))
    
```

```python
story_names = ['11', '12', '13', '14', '21', '22', '23', '24', '31', '32', '33', '34', '41', '42', '43', '44']

for i in range(len(dfs)):
    dfs[i]['audio'] = dfs[i]['audio'].map(lambda x: str(x)[:-4])
    dfs[i]['audio'] = dfs[i]['audio'].apply(lambda x: "{}{}".format(x, 'mp3'))
    new_dfs = pd.DataFrame(dfs[i]) 
    new_dfs.to_excel(r'/Users/alexreblando/Documents/GitHub/ebs/behavioral_August2019/psychopy experiment/audio_excel_sheets_mp3/' + story_names[i] +'.xlsx')
    
```
